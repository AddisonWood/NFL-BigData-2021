```python
plays["targetReceiver"] = np.NaN
plays["coveringBack"] = np.NaN

for i in range(len(plays)):
    if ("to" in plays["playDescription"][i]):
        cut1 = plays["playDescription"][i][plays["playDescription"][i].find("to")+3:]
        cut2 = cut1[:cut1.find(" ")]
        if (cut2[-1] == '.'):
            plays["targetReceiver"][i] = cut2[:-1]
        else:
            plays["targetReceiver"][i] = cut2
        
    
plays["targetReceiver"]
```
